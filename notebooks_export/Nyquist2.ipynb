### Nyquistkriteriet

Följande exwmpel illustrerar hur de olika delarna av *Nyquist kontur* som omsluter HHP avbildas. Speciellt hur den lilla halvcirkeln vid origo avbildas beroende på hur många poler det öppna systemet har i origo.

Vi skriver det öppna systemet så att antalet poler i origo blir explicit, $k$ stycken
$$
    G(s) = \frac{1}{s^k}\tilde{G}(s),
$$
vilket betyder att $\tilde{G}(0)$ är ett reelt tal skillt från noll, positivt eller negativt.

Vi väljer som exempelsystem
$$
    \tilde{G}(s) = \frac{10}{(s-1)}
$$
som har en pol i HHP.  Att polen ligger i HHP spelar ingen roll i examplet, du kan flytta den och se hur resultatet ändras.

Vi ska nu studera hur $G(\gamma)$ avbildas
då $\gamma$ definieras av *Nyquists kontur*, bestående av fyra segment:
\begin{align*}
    I:\quad &s=i\omega, \quad\omega=r\to\infty       &(r\to 0) \\
    II:\quad &s=Re^{i\theta}, \quad\theta=\pi/2 \to -\pi/2    &(R\to\infty) \\
    III:\quad &s=i\omega, \quad\omega=-\infty\to -r  &(r\to 0) \\
    IV:\quad &s=re^{i\theta}, \quad\theta=-\pi/2 \to \pi/2 &(r\to 0)
\end{align*}

Följande sammanfattar hur $L(\gamma)$ ser ut:

\begin{align*}
    I:\quad &G(i\omega) \text{ går från }G(0^+) \text{ till } 0. \ G(0^+)=\tilde{G}(0) \text{ ifall } k=0.\ \text{ Om } k>0 \text{ så ligger } G(0^+) \\ & \text{ i oändligheten längs någon av koordinataxlarna. Vilken av dessa beror på antalet}\\ &\text{poler och tecken på } G(0^+). \text{ Vägen in till origo kan medföra incrklingar av -1.}\\
    II:  \quad & \text{Halvcirkeln "kollapsar" i origo. Detta gäller för alla system som har fler poler än nollställen},\\ & G(Re^{i\theta})\to 0, R\to\infty \\
    III:\quad &\text{Kurvan blir spegelbilden av den i segment }I \\
    IV:\quad &G(re^{i\theta})\approx \frac{\tilde{G}(0)}{r^ke^{i\,k\theta}}=\frac{\tilde{G}(0)}{r^k}e^{-i\,k\theta}, \text{ som går \textit{medurs} från }\frac{\tilde{G}(0)}{r^k}i^k \text{ till } \frac{\tilde{G}(0)}{r^k}i^k(-1)^k. \text{ Detta betyder att}\\ & \text{varje pol ger en teckenväxling, dvs en halvcirkel.}
\end{align*}

Låt oss titta på fallen ingen pol och en pol. Med en pol blir det en halvcirkel, den pligger i oändligheten men plottas för ett relativt litet  $1/r$ (dvs $r$ är inte så litet). Halvcirkeln borde heller inte börja vid imaginära axeln.

```python
import control as ct
import matplotlib as mpl
import matplotlib.pyplot as plt

G = ct.tf([10],[-1, 1],name='DCgain 1')
G2 = ct.tf([10],[-1, 1,0],name='1-integration')

count = ct.nyquist_plot([G,G2])
print(count)
```

```python
import control as ct
import matplotlib as mpl
import matplotlib.pyplot as plt

G2 = ct.tf([10],[-1, 1,0,0],name='1-integration')

count = ct.nyquist_plot([G2])
print(count)
```
