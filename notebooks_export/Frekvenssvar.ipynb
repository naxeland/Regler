### Frekvenssvar för några enkla överföringsfunktioner

#### Första ordningens system

Ett första ordningens system karakteriseras av förstärkningen $K$ och tidskonstanten $T$: 

$$
G(s) = \frac{K}{1+sT}
$$

Bode- och Nyquistdiagram visas nedan för följande tre exempel:

$$
G_1 = \frac{1}{s+1}, \quad G_2 = \frac{1}{1+2s}, \quad G_3 = \frac{0.5}{s+1}
$$

```python
import control as ct
import matplotlib as mpl
import matplotlib.pyplot as plt

G1 = ct.tf([1],[1, 1])    
G2 = ct.tf([1],[2, 1])    
G3 = ct.tf([0.5],[1, 1])  

ct.bode([G1, G2, G3])
ct.bode(ct.tf([1],[1]), omega_limits=[0.1, 1], linestyle='--', linewidth=1, color='b')
ct.bode(ct.tf([1],[1, 0]), omega_limits=[1, 10], linestyle='--', linewidth=1, color='b')
plt.legend([r'$G_1=\frac{1}{s+1}$', r'$G_2=\frac{1}{1+2s}$', r'$G_3=\frac{0.5}{s+1}$'])
plt.show()

count = ct.nyquist_plot([G1, G2, G3], omega_limits=[0.001,100])
plt.axis('equal')
plt.show()
```

#### Andra ordingens system med komplexa poler

Ett andra ordningens system karakteriseras av *förstärkningen* $K$, den *relativa dämpningen* $\zeta$ och den *odämpade egenfrekvensen* $\omega$: 

$$
G(s) = \frac{K}{s^2 + 2\zeta\omega + \omega^2}
$$

Exempel:

$$
G_4 = \frac{1}{s^2+s+1}, \quad G_5 = \frac{4}{s^2+2s+4}, \quad G_6 = \frac{1}{s^2+0.4s+1}
$$

```python
import numpy as np
import control as ct
import matplotlib as mpl
import matplotlib.pyplot as plt

G4 = ct.tf([1], [1, 1, 1])   # zeta = 0.5, w = 1
G5 = ct.tf([4], [1, 2, 4])   # zeta = 0.5, w = 2
G6 = ct.tf([1], [1, 0.4, 1])   # zeta = 0.2, w = 1

ct.bode([G4, G5, G6])
ct.bode(ct.tf([1],[1]), omega_limits=[0.1, 1], linestyle='--', linewidth=1, color='b')
ct.bode(ct.tf([1],[1, 0, 0]), omega_limits=[1, 10], linestyle='--', linewidth=1, color='b')
plt.legend([r'$G_4=\frac{1}{s^2+s+1}$', r'$G_5=\frac{4}{s^2+2s+4}$', r'$G_6=\frac{1}{s^2+0.4s+1}$'])
plt.show()

count = ct.nyquist_plot([G4, G5, G6], omega_limits=[0.001,100])
plt.axis('equal')
plt.show()
```
